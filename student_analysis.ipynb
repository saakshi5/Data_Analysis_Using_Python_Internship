**Importing Libraries**

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

**Loading the Data**

```python
df=pd.read_csv("student-mat.csv",sep=';')
```

```python
df.head()
```

**Dataset Shape**

```python
df.shape
```

**Fixing Data Types**

```python
df['age'] = df['age'].astype(int)
df['G1'] = df['G1'].astype(int)
df['G2'] = df['G2'].astype(int)
df['G3'] = df['G3'].astype(int)
```

**Dataset info & data types**

```python
df.info()
```

**Dataset Summary statistics**

```python
df.describe()
```

**Data Cleaning**

1.Checking missing values

```python
df.isnull().sum()
```

2. Removing duplicates

```python
df = df.drop_duplicates()
df.shape
```

**Data Analysis**

1. Average Final Grade

```python
print("Average Final Grades:\n",df['G3'].mean())
```

2.How Many Students Scored Above 15?

```python
students_above_15 = df[df['G3'] > 15].shape[0]
print("Students whos scored above 15 are",students_above_15)
```

3.Is Study Time Correlated with Performance?

```python
correlation = df['studytime'].corr(df['G3'])
print(correlation)
print("No relation as value close to 0")
```

4.Which Gender Performs Better on Average?

```python
df.groupby('sex')['G3'].mean()
```

**Data Visualization**

1.Histogram â€“ Final Grades

```python
plt.hist(df['G3'], bins=10)
plt.xlabel('Final Grade (G3)')
plt.ylabel('Number of Students')
plt.title('Distribution of Final Grades')
plt.show()
```

Above histogram states how grades are distributed, they are between 5.0 to 20.0 grades.

2.Bar Chart â€“ Gender vs Performance

```python
df.groupby('sex')['G3'].mean().plot(kind='bar')
plt.title('Average Final Grade by Gender')
plt.ylabel('Average Grade')
plt.show()
```

According to above Gender Vs Performance bar chart, boys has scored significantly more than girls.

3.Scatterplot: study time vs grades.

```python
plt.scatter(df['studytime'], df['G3'])
plt.xlabel('Study Time')
plt.ylabel('Final Grade (G3)')
plt.title('Study Time vs Final Grade')
plt.show()
```

A scatter plot was used to analyze the relationship between study time and final grades.The plot shows a weak to moderate positive trend. Students with higher study time tend to score slightly better, though performance varies
